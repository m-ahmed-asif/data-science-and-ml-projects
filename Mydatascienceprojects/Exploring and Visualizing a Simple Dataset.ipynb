```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Optional: Display plots inline if running in a notebook
%matplotlib inline
```

```python
# Load built-in Iris dataset from seaborn
iris = sns.load_dataset('iris')
iris
```

```python
# Shape of the dataset
print("Shape of dataset:", iris.shape)

# Column names
print("Columns:", iris.columns.tolist())

# First 5 rows of the dataset
print("\nFirst 5 rows:")
print(iris.head())
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
plt.hist(iris['petal_width'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```
